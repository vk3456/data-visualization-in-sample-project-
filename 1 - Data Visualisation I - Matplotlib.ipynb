After installating, let's import pyplot from matplotlib as plt (by convention) and numpy as np:

```python
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
x = np.linspace(0,5,21)
x
```

```python
y1 = 2**x
y2 = 5**x
```

## Simple line plot

```python
plt.plot(x,y1)
plt.xlabel('Effort',fontsize=15)
plt.ylabel('Earnings',fontsize=15)
plt.title('Life',fontsize=15)
#plt.show()   if not on JupyterNotebook, it's necessary to show the plot
```

## Multiplots on same canvas

```python
plt.subplot(1,2,1)  # rows,columns,which graph
plt.plot(x,y1,'r')

plt.subplot(1,2,2) 
plt.plot(x,y2,'b')
```

## Object oriented plotting
Involves instantiating a plot object and then calling methods and attributes of that object to modify the plots properties.

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1]) #left, bottom, length, height

axes.plot(x,y1)
axes.set_xlabel('Effort',fontsize=15)
axes.set_ylabel('Earnings',fontsize=15)
axes.set_title('Life',fontsize=15)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])  #left, bottom, length, height
axes2 = fig.add_axes([0.12,0.6,0.4,0.3])
#axes3 = fig.add_axes([1,0,1,1])

axes1.plot(x,y2,'r')
axes1.set_xlabel('effort1',fontsize=15)
axes1.set_ylabel('earnings',fontsize=15)
axes1.set_title('Real Life',fontsize=15)

axes2.plot(x,y1,'b')
axes2.set_xlabel('effort2')
axes2.set_ylabel('earnings')
axes2.set_title('Dream Life')
```

```python
print(axes2)
```

## Multiplots (OOP)

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(x,y1,'r')
axes[0].set_xlabel('effort',fontsize=15)
axes[0].set_ylabel('earnings',fontsize=15)
axes[0].set_title('Life 1',fontsize=15)

axes[1].plot(x,y2,'b')
axes[1].set_title('Life 2',fontsize=15)

fig.tight_layout()
```

```python
for item in axes:
    print(item)
```

## Figure Size

```python
fig = plt.figure(figsize=(4,3))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y1)
```

```python
fig, axes = plt.subplots(1,2,figsize=(8,4))

axes[0].plot(x,y1)
axes[1].plot(x,y2)

fig.tight_layout()
```

```python
for item in axes:
    print(item)
```

## Adding legends

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y1,label='level1')
axes.plot(x,y2,label='level2')

axes.legend(loc='upper left')
```

## Customize Figures

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y1,color='red',linestyle='--',
         linewidth=1,alpha=0.5,
         marker='*',markersize=30,
         markerfacecolor='yellow',
         markeredgecolor='blue',
         markeredgewidth=5)             # color RGB hex code
                                        # alpha --> tranparency

axes.set_xlim([0,3])
```

# Plot types

### scatter plot

```python
import pandas as pd
from pandas import Series,DataFrame
```

```python
df = pd.read_csv('data_car_cross.csv',index_col=0)

df.head()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('Horse Power')
axes.set_ylabel('Weight')
axes.scatter(df['horse power'],df['weight'])
```

### histogram plot

```python
plt.hist(df['cylinder'])
plt.xlabel('Number of cylinders')
plt.ylabel('Count')
```

### boxplot

```python
np_array = np.array(df[['mpg','acceleration']])

plt.boxplot(np_array)
plt.xticks([1,2],['Miles per gallon','acceleration'])
```
